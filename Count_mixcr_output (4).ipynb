**Caculate the V gene mismatches** 



```python
V= df['allVAlignments'] = df['allVAlignments'].astype(str)
print(V)
```

```python
#Substractions
S_V = V.str.count('S')

#Deletions
D_V = V.str.count('D')

#Insertions
I_V = V.str.count('I')

#mismatches
R_V= S_V+D_V+I_V
print(R_V)
```

**Count the number of mapped nucleotides**

```python
split= df['allVAlignments'].str.split('|', expand=True)
df1=split

print(split)
```

```python
df1.dtypes
```

```python
df1
```

```python
df2= df1.iloc[:,3]
df3= df1.iloc[:,4]

df4= pd.concat([df2, df3], axis=1)

df4
```
